```python
# initialization
import pandas as pd
import numpy as np
```

```python
# many different python collection and datafile types can be rendered into pandas dataframes
# Here we use assignment statements to populate a dictionary.
# The keys are the column names while lists are ordered collections of the actual data
walrus_data = dict()

walrus_data['name'] = ['Anji', 'ET', 'Jocko', 'Blanca', 'Meika']
walrus_data['gender'] = ['Female','Male','Male', 'Female','Female']
walrus_data['weight'] = [2000,3400,3200,2500,3000]
walrus_data['vocalization'] = ['Lound Grunt','piErCiNG WHistle','soft grunt','chortle','Chortle']

print(walrus_data)
```

```python
# now let's render the dictionary into a pandas dataframe
walruses = pd.DataFrame(walrus_data)
print(walruses)
```

```python
#Let's explore the dataframe using attributes and methods
#print(walruses.head()) #pandas method
#print(type(walruses)) #buit-in python function
print(walruses.shape) #attribute
```

```python
# You can chain Python methods and attributes together
# After each, you must know what the collection or datatype is, then chain a method or attribute
# of the output collection or datatype
print(walruses.head(2).shape) # here we specify how many records to return
```

```python
# pandas dataframes have their own methods and attributes, included to make your life easier
# when working with datasets brought in as dataframes.
walruses = walruses.sort_values('weight')
```

```python
# note that the investigator has used an assignment statement to update the variable's value ---
# that is, the dataframe, which is now sorted in ascending values of weight.
# if the researcher had simply written:
# walruses.sort_values('weight')
# then by design, a copy of the dataframe would be written out to STDOUT.
# python programmers often use assignment statements to update the value of a variable.
```

```python
print(walruses) 
```

```python
walruses.sort_values('weight') # the dataframe written to STDOUT
```

```python
# however, it is important to note that many pandas methods allow you to update a dataframe inplace.
# the named argument keyword in these methods is 'inplace', and its value by default is False.
# We can sort the walruses dataframe by setting inplace=True .'''
```

```python
walruses.sort_values('weight', ascending=False, inplace=True)
print(walruses)
```

```python
# We can extract a single column
# (What's really returned here is a pandas Series)
#print(walruses['weight'])
print(walruses['name'])

#print(walruses['name'].values)
```

## Vectorized Tests/Operations

```python
# Adding 1 to the 'weight' column all at once
walruses['weight']=walruses['weight']+1
```

```python
# Creating a new pandas Series to record how many pounds of food the walruses ate today
walrus_food_consumption = pd.Series([60,150,100,200], index=['Anji','ET','Jocko', 'Blanca'])
walrus_food_consumption
```

```python
# here, we add the series as a column to our dataframe
walruses['food_consumption'] = walrus_food_consumption
print(walruses)
```

```python
# the not-a-number datatype (nan) is defined in the IEEE754 specification for handling floating-point numbers.
# pandas fills empty [row, column] positions with nan's; and you can decide how to handle the nan's.
# NOTE: the nan changes the column items datatype to float64...
print(type(walruses['food_consumption'][0]))
print("\n") # newline inserted to type and dataset printouts below
# here, because the column holds numeric data (and perhaps because meika didn't eat),
# the researcher decides to replace the nan with a zero ---
walruses = walruses.fillna(0)
print(walruses)
```

```python
# Find walrus weight today, append values as a new column to our dataframe
# we can coerce the floats into ints
walruses['weight_today'] = walruses['weight'] + walruses['food_consumption'].astype('int')
print(walruses)
```

## Boolean Selection

```python
# Identify female walruses
femalewalruses = walruses['gender'] == 'Female'
print(femalewalruses)
```

```python
# return records of all female walruses
walruses[femalewalruses]
```

```python
# Identify female walruses with weight >=1000
# note the logical AND operator and the grouping parentheses below
bigfemalewalruses = (walruses['gender']=='Female') & (walruses['weight'] >=1000)
walruses[bigfemalewalruses]
```

## String methods for pandas

```python
walruses['vocalization'].lower()
# ^^^ this doesn't work because we're trying to apply a string method to a pandas series
# rather, we want to apply the method to the items in the series
```

```python
walruses['vocalization'] = walruses['vocalization'].str.lower()
# these chained methods ^^^ tell the interpreter to apply the str.lower() method
# to the string items in the series
# the assignment statement updates the column vector in the dataframe
```
